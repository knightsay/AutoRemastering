## Trim Videos

yt-dlp sometimes have extra black scenes - trim it

```python
!ffmpeg -ss 00:00:00 -to 05:42:33 -i datasets/raw_data/halo2_orig_720.mp4 \
    -c copy datasets/halo2_orig_trimmed_720.mp4
!rm datasets/raw_data/halo2_orig_720.mp4
```

```python
!ffmpeg -ss 00:00:00 -to 05:19:42 -i datasets/raw_data/halo2_rmst_720.mp4 \
    -c copy datasets/halo2_rmst_trimmed_720.mp4
!rm datasets/raw_data/halo2_rmst_720.mp4
```
